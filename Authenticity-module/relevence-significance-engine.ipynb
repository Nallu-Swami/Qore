```python
import torch
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Store both sorted and unsorted results
unsorted_results = []

for answer in answers:
    input_text = f"{question}{answer}"
    inputs = tokenizer(input_text, max_length=128, add_special_tokens=False, return_tensors='pt')
    inputs = {key: val.to(device) for key, val in inputs.items()}

    with torch.inference_mode():
        logits = model(**inputs).logits
        probas = torch.sigmoid(logits)[0].cpu().detach().numpy()

    relevance, specificity = probas
    confidence = (relevance + specificity) / 2  # Confidence score

    unsorted_results.append({
        "answer": answer, 
        "relevance": relevance, 
        "specificity": specificity, 
        "confidence": confidence
    })

# Create a sorted version of results
sorted_results = sorted(unsorted_results, key=lambda x: (x["relevance"], x["specificity"]), reverse=True)

# Process last column of DataFrame
last_column_list = df[df.columns[-1]].tolist()[1:]

# Print Unsorted Results
print("===== Unsorted Results =====")
for res, val in zip(unsorted_results, last_column_list):
    print(res['answer'])
    print(f"R: {res['relevance']:.2f}, S: {res['specificity']:.2f}, C: {res['confidence']:.2f}")
    print(f"Last Column Value: {val}\n")
```
